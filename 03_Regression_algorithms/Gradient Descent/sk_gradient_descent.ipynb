```python
class GDregressor:
    def __init__(self,m=-10,b=-10,lr=0.01,epochs=500):
        self.m=m
        self.b=b
        self.lr=lr
        self.epochs=epochs
        self.weights = None
        self.bias = None
        model_info = {
            'Slope' : self.m,
            'Intercept' : self.b,
            'Learning_rate' : self.lr,
            'Epochs' : self.epochs
        }
        print(model_info)

    def LinearRegression_fit(self,x,y,verbose=False):

        for i in range(self.epochs):
            loss_slope_b = -2 * np.sum( y - (self.m*X.ravel()) - self.b )
            loss_slope_m = -2 * np.sum((y - (self.m*X.ravel()) - self.b) * X.ravel())

            step_size_m =  self.lr * loss_slope_m
            step_size_b = self.lr * loss_slope_b

            self.b = self.b - step_size_b
            self.m = self.m - step_size_m

            if verbose:
                print(f'm = {self.m}')
                print(f'b = {self.b}')
        # Loop Ended Here


        model_data = {
            'Slope' : self.m,
            'Intercept' : self.b,
        }
        print(f'Linear Regression Model Trained! {model_data}')


    def LinearRegression_predict(self,x):
        return self.m*x + self.b

    def MultipleRegression_fit(self, X, y):
        X = np.array(X)
        y = np.array(y)

        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for epoch in range(self.epochs):
            # Prediction
            y_pred = np.dot(X, self.weights) + self.bias

            # Errors
            error = y_pred - y

            # Gradients
            dw = (2 / n_samples) * np.dot(X.T, error)
            db = (2 / n_samples) * np.sum(error)

            # Update
            self.weights -= self.lr * dw
            self.bias -= self.lr * db

    def MultipleRegression_predict(self, X):
        X = np.array(X)
        return np.dot(X, self.weights) + self.bias

    def MultipleRegression_get_params(self):
        return self.weights, self.bias
```

```python
import numpy as np

class GdRegressor:
    def __init__(self, learning_rate=0.01, epochs=1000):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y)

        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for epoch in range(self.epochs):
            # Prediction
            y_pred = np.dot(X, self.weights) + self.bias

            # Errors
            error = y_pred - y

            # Gradients
            dw = (2 / n_samples) * np.dot(X.T, error)
            db = (2 / n_samples) * np.sum(error)

            # Update
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

    def predict(self, X):
        X = np.array(X)
        return np.dot(X, self.weights) + self.bias

    def get_params(self):
        return self.weights, self.bias
```
