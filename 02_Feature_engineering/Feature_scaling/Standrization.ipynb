# Standardization

```python
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt
import seaborn as sns# visualization

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.head()
```

```python
df=df.iloc[:,2:]
df.head()
```

```python
df.shape
```

```python
df.describe()
```

# Standard Scaler

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop('Purchased', axis=1), df['Purchased'], test_size=0.2)

x_train.shape
```

```python
sl = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
sl.fit(x_train)
```

```python
scaler_train_x = sl.transform(x_train)
scaler_test_x = sl.transform(x_test)
```

```python
scaler_train_x
```

```python
df_x = pd.DataFrame(scaler_train_x, columns=x_train.columns)
df_x2 = pd.DataFrame(scaler_test_x, columns=x_test.columns)
df_x.shape
```

```python
df2= pd.concat([df_x,df_x2],axis=0)
df2.shape
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(10,4))

sns.scatterplot(x=x_train['Age'],y=x_train['EstimatedSalary'], ax=ax[0])
ax[0].set_title('Before scaling')

sns.scatterplot(x=df_x['Age'],y=df_x['EstimatedSalary'], ax=ax[1], color='red')
ax[1].set_title('After scaling')
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(10,4))

sns.kdeplot(x=x_train['Age'], ax=ax[0])
sns.kdeplot(x=x_train['EstimatedSalary'], ax=ax[0])
ax[0].set_title('Before scaling')

sns.kdeplot(x=df_x['Age'], ax=ax[1])
sns.kdeplot(x=df_x['EstimatedSalary'], ax=ax[1])
ax[1].set_title('After scaling')
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(10,4))

sns.kdeplot(x=x_train['Age'], ax=ax[0])
ax[0].set_title('Before scaling')

sns.kdeplot(x=df_x['Age'], ax=ax[1])
ax[1].set_title('After scaling')
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(10,4))

sns.kdeplot(x=x_train['EstimatedSalary'], ax=ax[0])
ax[0].set_title('Before scaling')

sns.kdeplot(x=df_x['EstimatedSalary'], ax=ax[1])
ax[1].set_title('After scaling')
```

# Why scaling is important

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr_scale = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
print("Training Accuracy:",lr.score(x_train,y_train))
```

```python
lr_scale.fit(df_x,y_train)
print("Training Accuracy:",lr_scale.score(df_x,y_train))
```

```python
print("Training Accuracy without scaling:",lr.score(x_train,y_train))
print("Training Accuracy with scaling:",lr_scale.score(df_x,y_train))
```
