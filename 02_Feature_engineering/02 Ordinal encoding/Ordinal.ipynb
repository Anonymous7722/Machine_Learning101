```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
# read csv
# see the data head()
# take usable columns
# train_test split
# Ordinal encoding input features
# Label encoding Target feature




df = pd.read_csv('customer.csv')
```

```python
df.head()
```

```python
df=df.iloc[:,2:]
df.head()
```

```python
df['education'].unique()
```

```python
x_train,x_test,y_train,y_test = train_test_split(df[['review','education']], df[['purchased']],test_size=0.2)
```

```python
Or = OrdinalEncoder(categories=[['Poor','Average','Good'], ['School','UG','PG']])
```

```python
Or.fit(x_train)
```

```python
train_or = Or.transform(x_train)
test_or = Or.transform(x_test)
```

```python
train_or = pd.DataFrame(train_or, columns=['review','education'])
test_or = pd.DataFrame(test_or, columns=['review','education'])
train_or.head()
```

```python
lo = LabelEncoder()
lo.fit(y_train)

train_lo = lo.transform(y_train)
test_lo = lo.transform(y_test)
train_lo
```

```python
train_lo = pd.DataFrame(train_lo, columns=['purchase'])
test_or = pd.DataFrame(test_or, columns=['purchase'])

train_lo
```
